```python
import gensim, logging, os, logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
class MySentences(object):
    def __init__(self, dirname):
        self.dirname = dirname
 
    def __iter__(self):
        for fname in os.listdir(self.dirname):
            for line in open(os.path.join(self.dirname, fname)):
                yield line.split()
```

```python
sentences = MySentences('/home/bahbbc/Documents/teste-personalidade') # a memory-friendly iterator
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = gensim.models.Word2Vec(sentences, workers=50)
model.save('/home/bahbbc/Documents/teste-personalidade/personality_1')
```

```python
print(model)
print(model.wv.vocab)
```

```python
print(model.similarity('hj', 'gato'))
```
